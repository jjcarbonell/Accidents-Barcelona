```python
import pandas as pd
import numpy as np
```

```python
datos = pd.read_csv('C:\master\dataset.csv')
```

```python
datosA = datos[(datos['Accident'] == 1)]
len(datosA)
```

```python
datosN = datos[(datos['Accident'] == 0)].sample(n=36517)
len(datosN)
```

```python
datosAN = datosA.append(datosN)
```

```python
len(datosAN)
```

```python
datosAN
```

```python
datosAN["Accident"].value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder
LE = LabelEncoder()
LE.fit(datosAN["Shift"])
```

```python
list(LE.classes_)
datosAN["Shift"] = LE.transform(datosAN["Shift"])
datosAN["Shift"].value_counts()
```

```python
X = datosAN.drop("Accident",1)
X = X.drop("date",1)
```

```python
y = datosAN["Accident"]
```

```python
print(X.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.30, random_state=1)
```

```python
print (X_train.shape)
print (X_test.shape)
print (y_train.shape)
print (y_test.shape)
```

```python
from sklearn.tree import DecisionTreeClassifier
DTC = DecisionTreeClassifier(criterion='entropy', random_state=1)
```

```python
DTC.fit(X_train, y_train)
```

```python
y_train_pred = DTC.predict_proba(X_train)
```

```python
from sklearn.metrics import roc_auc_score
```

```python
print(roc_auc_score(y_train, y_train_pred[:, 1]), "Árbol de decisión - Datos de entrenamiento")
```

```python
y_test_pred = DTC.predict_proba(X_test)
```

```python
print(roc_auc_score(y_test, y_test_pred[:, 1]), "Árbol de decisión - Datos de test")
```
